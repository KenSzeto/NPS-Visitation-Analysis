## 03. Data Consistency Checks

```python
df_nps.info()
```

```python
#checking for mixed data columns

for col in df_nps.columns.tolist():
  weird = (df_nps[[col]].applymap(type) != df_nps[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_nps[weird]) > 0:
    print (col)
```
